```python
def clean_activity(activity):
    activity = str(activity).lower()
    
    if 'surf' in activity:
        return 'Surfing'
    elif 'swim' in activity or 'bathing' in activity or 'floating' in activity:
        return 'Swimming'
    elif 'wading' in activity or 'standing' in activity or 'treading' in activity:
        return 'Shallow water'
    elif 'div' in activity:
        return 'Diving'
    elif 'fish' in activity:
        return 'Fishing'
    elif 'snorkel' in activity:
        return 'Snorkeling'
    elif 'board' in activity:
        return 'Board sports'
    else:
        return 'Other'
```

```python
df['activity_clean'] = df['activity'].apply(clean_activity)
```

# Comparativa de datos de ataques pais y actividad

```python
# Top 10 países con más ataques
top_paises = df['country'].value_counts().head(10).index

# Top 5 actividades más frecuentes
top_actividades = df['activity'].value_counts().head(5).index

# Crosstab filtrado
cross = pd.crosstab(df['country'], df['activity'])
cross_filtrado = cross.loc[top_paises, top_actividades]

print("Tabla crosstab - Top 10 países x Top 5 actividades:")
print(cross_filtrado)
```

```python
cross_pct = cross_filtrado.div(cross_filtrado.sum(axis=1), axis=0) * 100

print("Porcentaje de actividades dentro de cada país:")
print(cross_pct.round(2))
```

```python
def clasificar_actividades_peligro(df):
    cross_tipo = pd.crosstab(df['activity'], df['type'])  # Cruza actividad con tipo de ataque

    provocados = cross_tipo['Provoked'] if 'Provoked' in cross_tipo.columns else 0  # Cuenta ataques provocados
    no_provocados = cross_tipo['Unprovoked'] if 'Unprovoked' in cross_tipo.columns else 0  # Cuenta ataques no provocados

    condiciones = no_provocados > provocados  # Si hay más no provocados, la actividad se considera peligrosa
    clasificacion = []

    for _, condicion in condiciones.items():
        if condicion:
            clasificacion.append("Peligrosa")
        else:
            clasificacion.append("Poco peligrosa")

    resultado = pd.DataFrame({
        'Actividad': cross_tipo.index,
        'Provocados': provocados.values if hasattr(provocados, 'values') else [provocados] * len(cross_tipo),
        'No_Provocados': no_provocados.values if hasattr(no_provocados, 'values') else [no_provocados] * len(cross_tipo),
        'Total': cross_tipo.sum(axis=1).values,
        'Clasificacion': clasificacion
    })  # Crea la tabla final con los resultados

    return resultado.sort_values('Total', ascending=False)  # Ordena de mayor a menor número total de ataques


resultado = clasificar_actividades_peligro(df)  # Ejecuta la función y guarda el resultado

for _, fila in resultado.head(10).iterrows():  # Recorre las 10 actividades con más ataques
    if fila['No_Provocados'] > fila['Provocados']:
        print(f"🔴 {fila['Actividad']} - Peligrosa")
    else:
        print(f"🟢 {fila['Actividad']} - Poco peligrosa")


```

## Porcentaje de peligrosidad 

```python
def porcentaje_peligrosidad(df):
    
    total_actividad = df.groupby('activity').size()
    no_prov_actividad = df[df['type'] == 'Unprovoked'].groupby('activity').size()

    total_pais = df.groupby('country').size()
    no_prov_pais = df[df['type'] == 'Unprovoked'].groupby('country').size()

    
    peligrosidad_actividad = (no_prov_actividad / total_actividad * 100).fillna(0).sort_values(ascending=False)
    peligrosidad_pais = (no_prov_pais / total_pais * 100).fillna(0).sort_values(ascending=False)

    df_actividad = peligrosidad_actividad.reset_index()
    df_actividad.columns = ['Actividad', 'Porcentaje_peligrosidad']

    df_pais = peligrosidad_pais.reset_index()
    df_pais.columns = ['Pais', 'Porcentaje_peligrosidad']

    return df_actividad, df_pais
```

```python
actividad_riesgo, pais_riesgo = porcentaje_peligrosidad(df)

print("Porcentaje de peligrosidad por actividad:")
print(actividad_riesgo.head(10))

print("\nPorcentaje de peligrosidad por país:")
print(pais_riesgo.head(10))
```

```python
def porcentaje_por_actividad_y_pais(df):
    total_global = len(df)

    actividad = df.groupby('activity').size().reset_index(name='Total_Ataques')
    actividad['Porcentaje'] = actividad['Total_Ataques'] / total_global * 100

    pais = df.groupby('country').size().reset_index(name='Total_Ataques')
    pais['Porcentaje'] = pais['Total_Ataques'] / total_global * 100

    actividad = actividad.sort_values('Porcentaje', ascending=False).reset_index(drop=True)
    pais = pais.sort_values('Porcentaje', ascending=False).reset_index(drop=True)

    return actividad, pais
```

```python
actividad_pct, pais_pct = porcentaje_por_actividad_y_pais(df)

print("Porcentaje de ataques por actividad:")
print(actividad_pct.head(10))

print("\nPorcentaje de ataques por país:")
print(pais_pct.head(10))
```
