```python
df = df[['country', 'activity', 'type', 'year', 'species ', 'fatal y/n']]
df.head()
```

```python
df = df.dropna(subset=['country', 'activity', 'type'])
```

## Revisión de duplicados

```python
df = df.drop_duplicates()
```

```python
df['activity'].head(10)
```

```python
def clean_activity(activity):
    activity = str(activity).lower()
    
    if 'surf' in activity:
        return 'Surfing'
    elif 'swim' in activity or 'bathing' in activity or 'floating' in activity:
        return 'Swimming'
    elif 'wading' in activity or 'standing' in activity or 'treading' in activity:
        return 'Shallow water'
    elif 'div' in activity:
        return 'Diving'
    elif 'fish' in activity:
        return 'Fishing'
    elif 'snorkel' in activity:
        return 'Snorkeling'
    elif 'board' in activity:
        return 'Board sports'
    else:
        return 'Other'
```

```python
df['activity_clean'] = df['activity'].apply(clean_activity)
```

```python
df['activity_clean'].value_counts()
```
